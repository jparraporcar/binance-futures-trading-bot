__Results: the benefits of having a more sensitive signal seems to outperform a slower one, in the sense that fast changes in price are captured by the faster EMA before than they are captured by the slower EMA, a compromise have to be found and the parameter can be adjusted depending in market conditions.
A TOO FAST ema_fast include noise in the calculation of the macd_diff. The ema_fast does not have to be lowered to capture big drops or increases in price since, in order to do this, a helper function to detect changes at the websocket desired granularity has to be used__
