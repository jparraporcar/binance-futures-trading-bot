```python
class Tester_macd_ls_backtester_sma():
    def __init__(self,df=None):
        self.df = df
        self.test_pass_smaconflong = True
        self.test_pass_macdconflong = True
        self.test_pass_smaconfshort = True
        self.test_pass_macdconfshort = True
        self.test_pass_all_long_or_short = True
        self.test_list = []
    
    def start_test_smaconflong(self):
        #ensuring that the long position (=1) is correctly assigned
        for index, data in self.df.iterrows():
            cond_long_partial_1 = (data.macd_inv_sign == 1 and data.sma_inv_sign == 1) # macd cross long + sma cross long
            cond_long_partial_2 = (data.macd_inv_sign == 1 and data.sma_conf == 1) # macd cross long + sma positive
            cond_long_total_1_and_2 = cond_long_partial_1 or cond_long_partial_2
            
            if (cond_long_total_1_and_2):
                if (data.position == 1):
                    pass
                else:
                    self.test_pass_smaconflong = False
                    break
        
        if (self.test_pass_smaconflong == False):
            self.test_list.append(['macd long cross + sma confirmation', 'FAILURE'])
        
        else:
            self.test_list.append(['macd long cross + sma confirmation', 'SUCCESS'])
    
    def start_test_macdconflong(self):
        for index, data in self.df.iterrows():
            cond_long_partial_3 = (data.sma_inv_sign == 1 and data.macd_inv_sign == 1)
            cond_long_partial_4 = (data.sma_inv_sign == 1 and data.macd_diff_conf == 1)
            cond_long_total_3_and_4 = cond_long_partial_3 or cond_long_partial_4

            if (cond_long_total_3_and_4):
                if (data.position == 1):
                    pass
                else:
                    self.test_pass_macdconflong = False
                    break
        
        if (self.test_pass_macdconflong == False):
            self.test_list.append(['sma long cross + macd confirmation', 'FAILURE'])
        
        else:
            self.test_list.append(['sma long cross + macd confirmation', 'SUCCESS'])
            
    def start_test_smaconfshort(self):
        #ensuring that the long position (=1) is correctly assigned
        for index, data in self.df.iterrows():
            cond_short_partial_1 = (data.macd_inv_sign == -1 and data.sma_inv_sign == -1) # macd cross long + sma cross long
            cond_short_partial_2 = (data.macd_inv_sign == -1 and data.sma_conf == -1) # macd cross long + sma positive
            cond_short_total_1_and_2 = cond_short_partial_1 or cond_short_partial_2
            
            if (cond_short_total_1_and_2):
                if (data.position == -1):
                    pass
                else:
                    self.test_pass_smaconfshort = False
                    break
        
        if (self.test_pass_smaconfshort == False):
            self.test_list.append(['macd short cross + sma confirmation', 'FAILURE'])
        
        else:
            self.test_list.append(['macd short cross + sma confirmation', 'SUCCESS'])    

    def start_test_macdconfshort(self):
        for index, data in self.df.iterrows():
            cond_short_partial_3 = (data.sma_inv_sign == -1 and data.macd_inv_sign == -1)
            cond_short_partial_4 = (data.sma_inv_sign == -1 and data.macd_diff_conf == -1)
            cond_short_total_3_and_4 = cond_short_partial_3 or cond_short_partial_4

            if (cond_short_total_3_and_4):
                if (data.position == -1):
                    pass
                else:
                    self.test_pass_macdconfshort = False
                    break
        
        if (self.test_pass_macdconfshort == False):
            self.test_list.append(['sma short cross + macd confirmation', 'FAILURE'])
        else:
            self.test_list.append(['sma short cross + macd confirmation', 'SUCCESS'])
    
    def start_test_all_long_or_short(self):
        for index, data in self.df.iterrows():
            if (data.position == 0):
                pass
            else:
                self.test_pass_all_long_or_short = False
        
        #actually it is success if position == 0 happens at the beginning of the dataframe since no conditions was met
        #initially
        if (self.test_pass_all_long_or_short == False):
            self.test_list.append(['all positions long or short', 'FAILURE'])
        else:
            self.test_list.append(['all positions long or short', 'SUCCESS'])

    def run_all_tests(self):
        self.start_test_smaconflong()
        self.start_test_macdconflong()
        self.start_test_smaconfshort()
        self.start_test_macdconfshort()
        self.start_test_all_long_or_short()
        for lst in self.test_list:
            print(lst)
```
