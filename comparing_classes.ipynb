```python
import matplotlib.dates as mdates
# from IPython.core.display import display, HTML
# display(HTML("<style>.container { width:100% !important; }</style>"))
colors=[]


fig, (close_ax, macd_ax) = plt.subplots(nrows=2, ncols=1, figsize=(30,20), gridspec_kw={'height_ratios': [4,2]}, sharex=True)

close_ax.grid(visible=True, which='major', axis='x', color='grey')
macd_ax.grid(visible=True, which='major', axis='x', color='grey')
close_ax.grid(visible=True, which='minor', axis='x', color='grey')
macd_ax.grid(visible=True, which='minor', axis='x', color='grey')
close_ax.grid(visible=True, which='major', axis='y', color='grey')
macd_ax.grid(visible=True, which='major', axis='y', color='grey')

close_ax.tick_params(labelrotation=45)
macd_ax.tick_params(labelrotation=45)

close_ax.margins(0)
macd_ax.margins(0)

days = mdates.DayLocator()
macd_ax.xaxis.set_minor_locator(days)

close_ax.set_ylim(auto=True)

# close_ax.plot(macd_peak_ee.data_init.index, macd_simp_ee.data_init.macd_log_returns_acum)
# close_ax.plot(macd_peak_ee.data_init.index, macd_simp_ee.data_init.multiple_hold_acum)
close_ax.plot(macd_peak_ee.data_init.index, macd_peak_ee.data_init.Close)

for index, value in macd_peak_ee.data_init.macd_diff.iteritems():
    if value > 0:
        colors.append('g')
    else:
        colors.append('r')

macd_ax.bar(x= macd_peak_ee.data_init.index, height= macd_peak_ee.data_init.macd_diff, width=0.035, align='center', color=colors)
date_form = mdates.DateFormatter('%Y-%m-%dT%H:%M:%S')
close_ax.xaxis.set_major_formatter(date_form)

macd_peak_ee_peak_neg = macd_peak_ee.data_init.loc[macd_peak_ee.data_init.peak == 1]
macd_peak_ee_peak_pos = macd_peak_ee.data_init.loc[macd_peak_ee.data_init.peak == -1]


close_ax.scatter(macd_peak_ee_peak_pos.index, macd_peak_ee_peak_pos.Close.loc[macd_peak_ee_peak_pos.index], marker='^', color='r', s=100)
close_ax.scatter(macd_peak_ee_peak_neg.index, macd_peak_ee_peak_neg.Close.loc[macd_peak_ee_peak_neg.index], marker='^', color='g', s=100)
# btcusdt_inv_sign_neg = btcusdt.loc[btcusdt.inv_sign == -1]
# btcusdt_inv_sign_pos = btcusdt.loc[btcusdt.inv_sign == 1]
# close_ax.scatter(btcusdt_inv_sign_neg.index, btcusdt_inv_sign_neg.Close.loc[btcusdt_inv_sign_neg.index], marker='^', color='r', s=100)
# close_ax.scatter(btcusdt_inv_sign_pos.index, btcusdt_inv_sign_pos.Close.loc[btcusdt_inv_sign_pos.index], marker='^', color='g', s=100)
```
